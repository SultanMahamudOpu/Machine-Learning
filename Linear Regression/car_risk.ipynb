# Car Risk Analysis

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('car driving risk analysis.csv')
df.head()
```

```python
x=df[['speed']]
x
```

```python
y=df[['risk']]
y
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
plt.scatter(x.mean(),y.mean())
plt.scatter(df['speed'],df['risk'])
plt.xlabel('speed')
plt.ylabel('risk')
plt.title("Speed Risk")
```

```python
reg.fit(x,y)
```

```python
m=reg.coef_
m
```

```python
c=reg.intercept_
c
```

```python
reg.predict([[13]])
```

```python
df['Predicted_y']=reg.predict(x)
df.head()
```

# Lose and Cost Funciton

```python
df['lose']=df['speed']-df['Predicted_y']
df.head()
```

# MSE and MAE

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mse=mean_squared_error(df['risk'],df['Predicted_y'])
mse
```

```python
mae=mean_absolute_error(df['risk'],df['Predicted_y'])
mae
```

# Accuracy

```python
reg.score(x,y)
```

# R^2 value

```python
from sklearn.metrics import r2_score
r2_score(y,reg.predict(x))
```
