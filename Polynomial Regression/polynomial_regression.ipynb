```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Import Linear dataset

```python
df=pd.read_csv('linear_data.csv')
df.head()
```

```python
x_linear=df[['x']]
y_linear=df[['y']]
```

# import non-linear dataset

```python
df2=pd.read_csv('nonlinear_data.csv')
df2.head()
```

```python
x_nonlinear=df2[['x']]
y_nonlinear=df2[['y']]
```

# Visual representation

```python
# linear
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(x_linear,y_linear,color='blue',label='Data with Linear Relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Relationship')
plt.legend()

# non-linear
plt.subplot(1,2,2)
plt.scatter(x_nonlinear,y_nonlinear,color='green',label='Data with non-linear Relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Non-linear Relationship')
plt.legend()

plt.tight_layout()
plt.show()
```

 # Linear

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_linear,y_linear)
```

```python
reg.score(x_linear,y_linear)
```

```python
plt.scatter(x_linear,y_linear,color='blue',label='Data with Linear Relationship')
plt.plot(x_linear,reg.predict(x_linear),color='red',label='Best fit line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Relationship')
plt.legend()
```

# Non-linear

```python
reg2=LinearRegression()
```

```python
reg2.fit(x_nonlinear,y_nonlinear)
```

```python
reg2.score(x_nonlinear,y_nonlinear)
```

```python
plt.scatter(x_nonlinear,y_nonlinear,color='green',label='Data with non linear relationship')
plt.plot(x_nonlinear,reg2.predict(x_nonlinear),color='red',label='Best fit line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Relationship')
plt.legend()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=6)
x_poly=poly.fit_transform(x_nonlinear)
```

```python
x_poly.shape
```

```python
reg_poly=LinearRegression()
```

```python
reg_poly.fit(x_poly,y_nonlinear)
```

```python
reg_poly.score(x_poly,y_nonlinear)
```

```python
plt.scatter(x_nonlinear,y_nonlinear,color='green',label='Data with non-linear relationship')
plt.plot(x_nonlinear,2*np.sin(x_nonlinear),color='orange',label='Trrue non-linear relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Non-linear Relationship')
plt.legend()
```

# Seperate Train Test

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x_poly,y_nonlinear,test_size=30,random_state=1)
```

```python
reg=LinearRegression()
```

```python
reg.fit(xtrain,ytrain)
```

testing score

```python
reg.score(xtest,ytest)
```

training score

```python
reg.score(xtrain,ytrain)
```
